```python
import pandas as pd
```

```python
# Create a function that rank different variables for each city.
#     The only parameter 'boolean' decides if we rank the variable in an ascending way or descending way.
#        - The rank in the descending way, shows the actual rank. The smaller actual ranking value, the better. This is in accordance with our common sense.  
#        - * The reversed rank (rank in the ascending way) will only be used in the radar graph. The larger value, the higher actual ranking value, the better.

def rank(boolean):
# import the merged_final.csv which contains the actual score of each variable, our rankings are based on that score.
    df = pd.read_csv('merged_final.csv')
    for col in df.columns:
        if col in ['environment_score', 'leisure+accessibility_score', 'cost_of_living_score', 'edu_score', 'job_score', 'safety_score']:
          # We select out the columns that will be ranked later
          # to protect the original score data from being changed, we create the copy of it, which is a list named 's'
            s = df[col].copy()

            # Here you can see the parameter 'boolean'
            # after this script, we can see the each column ranked separately, with the score in each row of tuple
            s.sort_values(ascending = boolean, inplace = True)
            rank = [i for i in range(len(s))]
            # in this for loop, we change the score into an order number
            for n in rank:
                s = s.replace(s.iloc[n], n+1)

            # Get a new name for the ranked column
            s.rename(s.name+'_rank', inplace=True)

            # we merge the ranked column with order numbers with the original table, using the index of each row
            df = pd.merge(df, s, left_index=True, right_index=True)
            df_rank = df.copy()

            # Here we drop the score columns because we don't want them anymore
            df_rank = df_rank.drop(['environment_score', 'leisure+accessibility_score', 'cost_of_living_score', 'edu_score', 'job_score', 'safety_score'], axis=1)

            # To make sure that the order numbers will come out as integers, here we formalize them using .astype()
            for col in df_rank.columns:
                if col in ['environment_score_rank', 'leisure+accessibility_score_rank',
                           'cost_of_living_score_rank', 'edu_score_rank', 'job_score_rank',
                           'safety_score_rank']:
                    df_rank[col] = df_rank[col].astype('int32')

            # we export the ranked table. It will be used in UI&Visualization part.
            df_rank.to_csv('ranked_index_ascending_'+str(boolean)+'.csv')
            
if __name__ == '__main__':
    rank(True)
    rank(False)
```
